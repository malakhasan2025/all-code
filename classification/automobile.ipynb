```python
import numpy as np
import pandas as pd
```

```python
data =pd.read_csv("Automobile_data.csv")
```

```python
data = data.drop(data.columns[8], axis=1)
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
impute.fit(data.iloc[:,1:2])
data.iloc[:,1:2]=impute.transform(data.iloc[:,1:2])
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
impute.fit(data.iloc[:,21:25])
data.iloc[:,21:25]=impute.transform(data.iloc[:,21:25])
```
