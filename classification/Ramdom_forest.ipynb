# Ramdom forest



# Importing Lib

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

# Splitting  the dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)

```

```python
print(y_test)
```

# feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Tranning the Random_forest model with tranning data


```python
from sklearn.ensemble import RandomForestClassifier
clasifier = RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=42)
clasifier.fit(X_train, y_train)
```

```python
print(clasifier.predict(sc.transform([[30,87000]])))
```

# predicting the test set result

```python
print(X_test)
y_pred=clasifier.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), axis=1))
```

# Making the Confusion matric


```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Visualising traning result

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

for i in range(1, 42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    model = DecisionTreeClassifier(random_state=0)
    model.fit(X_train, y_train)
    print("Random state:", i, "Accuracy:", model.score(X_test, y_test))
```
